```python
print("Resolvendo um sistema linear")
""""
    2x + 6y = 6
    5x + 3y = -9
    x = ? e y = ?
    Matriz de coeficientes e variáveis dependentes
    [2 6 = [6
    5 3 ] = -9]
"""

coef = np.array([[2,6], [5,3]])
vardeps = np.array([6, -9])
print(coef)
print(vardeps)

solucao = np.linalg.solve(coef, vardeps)
print(solucao)
print(f" X = {solucao[0]}")
print(f" Y = {solucao[1]}")
```

```python
import numpy as np  # Importa a biblioteca NumPy, que fornece funções matemáticas e de álgebra linear.

print("Resolvendo um sistema linear de 3 variáveis com Numpy")
"""
Sistema linear: x = euro, y = dólar e z = iene
2x + 5y + 2z = 18
5x + 3y + 4z = 25
3x + 3y + 4z = 19
x = ?
y = ?
z = ?
"""

# Define a matriz de coeficientes do sistema linear.
# Cada linha representa os coeficientes de x, y, e z em cada uma das equações.
coef = np.array(
                [[2, 5, 2],  # Coeficientes da primeira equação: 2x + 5y + 2z
                 [5, 3, 4],  # Coeficientes da segunda equação: 5x + 3y + 4z
                 [3, 3, 4]]  # Coeficientes da terceira equação: 3x + 3y + 4z
                )

# Define o vetor de termos independentes (resultados das equações).
# Esses são os valores à direita do sinal de igual nas equações.
vardeps = np.array([18, 25, 19])

# Imprime a matriz de coeficientes para verificação.
print(coef)

# Imprime o vetor de termos independentes para verificação.
print(vardeps)

# Resolve o sistema de equações lineares usando a função linalg.solve() do NumPy.
# A função retorna um array com os valores das variáveis x, y, z que satisfazem o sistema.
solucao = np.linalg.solve(coef, vardeps)

# Imprime a solução completa, que é um array contendo os valores de x, y, z.
print(solucao)

# Imprime os valores individuais de x, y, e z de forma mais legível.
print(f"x = {solucao[0]}")  # Valor de x
print(f"y = {solucao[1]}")  # Valor de y
print(f"z = {solucao[2]}")  # Valor de z
```

```python
# Importa a biblioteca matplotlib.pyplot para criar gráficos
import matplotlib.pyplot as plt

# Define dois conjuntos de dados para plotagem
data_1 = [10, 5, 2, 4, 6, 8]  # Primeiro conjunto de dados
data_2 = [1, 2, 4, 8, 7, 4]   # Segundo conjunto de dados

# Cria os valores do eixo X, multiplicando por 10 para espaçar os pontos
# np.array(range(len(data_1))) cria um array [0, 1, 2, 3, 4, 5]
# Multiplicando por 10, obtemos [0, 10, 20, 30, 40, 50]
x = 10 * np.array(range(len(data_1)))

# Plota o primeiro conjunto de dados (data_1) como pontos verdes ('go')
# 'g' indica a cor verde e 'o' indica que os pontos serão plotados como círculos
plt.plot(x, data_1, 'go')

# Plota o mesmo conjunto de dados (data_1) como uma linha tracejada vermelha
# 'k:' originalmente indica uma linha tracejada preta, mas a cor é sobrescrita por 'color = "red"'
plt.plot(x, data_1, 'k:', color='red')

# Plota o segundo conjunto de dados (data_2) como uma linha tracejada azul
# 'k--' originalmente indica uma linha tracejada preta, mas a cor é sobrescrita por 'color = "blue"'
plt.plot(x, data_2, 'k--', color='blue')

# Exibe o gráfico gerado
plt.show()
```
