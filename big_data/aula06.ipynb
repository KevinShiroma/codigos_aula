```python
dados = pd.Series([-70, -90, 130, 150], index=['jan', 'fev', 'mar', 'abr'])
cor = []
for num in dados.values:
    if num > 0:
        cor.append("green")
    else:
        cor.append("red")
plt.bar(dados.index, dados.values, color=cor)
plt.title("Vendas Quadrimestre - 2022")
plt.ylabel("Vendas em Milhares (R$)")
plt.xlabel("1Â° Quadrimestre 2022")
plt.grid(True)
plt.show()
```

```python
dados = pd.Series([70, 90, 130, 150], index=['jan', 'fev', 'mar', 'abr'])
cor = []
for num in dados.values:
    if num > 100:
        cor.append("green")
    else:
        cor.append("red")
plt.pie(dados.values, labels = dados.index, colors=cor)
plt.title("Vendas Quadrimestre - 2022")
plt.ylabel("Vendas em Milhares (R$)")
plt.xlabel("1Â° Quadrimestre 2022")
plt.grid(True)
plt.show()
```

```python
pop_dict = {"SÃ£o Paulo": 46649132,
            "Minas Gerais": 21411923,
            "Rio de Janeiro": 17463349,
            "Bahia": 14985284}

area_dict =  {"SÃ£o Paulo": 248219,
            "Minas Gerais": 586521,
            "Rio de Janeiro": 43750,
            "Bahia": 564760}

pop_area = pd.Series(area_dict)
pop_serie = pd.Series(pop_dict)
print(pop_serie)
print(pop_serie.index)
print(pop_serie.values)

df = pd.DataFrame({"populaÃ§Ã£o": pop_serie, "area": pop_area})

df["densidade"] = df["populaÃ§Ã£o"] / df["area"]

display(df)
```
